```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
```

# Exercise 4: Burgers Equation
$$u(x) = \sin(\frac{\pi x}{L})+\epsilon$$

```python
L = 1000
dt = 0.1
N = 1000
Nt = 10000

x = np.linspace(0,L,N)
dx = x[1] - x[0]
eps = 0.5

u = np.zeros((Nt, N))
u[0] = np.sin(np.pi*x/L)+eps
```

## Left-Hand

```python
A = np.diag(np.ones(N))
A += np.diag(-np.ones(N-1), -1)
A[0,-1] = -1

for i in range(Nt-1):
    u[i+1] = u[i] -  dt/dx*u[i]*(A@u[i])

plt.imshow(u, aspect="auto", cmap="magma")
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

line, = ax.plot(u[0])

def animate(i):
    line.set_ydata(u[i])
    return line,

anim = animation.FuncAnimation(fig, animate, frames=Nt, interval=1)
anim.save('burgers_eq.gif') 
```

```python
plt.plot(u[-1])
```

## Right-Hand

```python
A = np.diag(np.ones(N))
A += np.diag(-np.ones(N-1), 1)
A[-1,0] = -1

for i in range(Nt-1):
    u[i+1] = u[i] -  dt/dx*u[i]*(A@u[i])

plt.imshow(u, aspect="auto", cmap="magma")
plt.show()
```

## Central

```python
A = np.diag(np.ones(N-1), 1)
A += np.diag(-np.ones(N-1), -1)
A[-1,0] = 1
A[0,-1] = -1

for i in range(Nt-1):
    u[i+1] = u[i] -  dt/(2*dx)*u[i]*(A@u[i])

plt.imshow(u, aspect="auto", cmap="magma")
plt.show()
```

```python
L = 1000
dt = 0.01
N = 1000
Nt = 1000

x = np.linspace(0,L,N)
dx = x[1] - x[0]
eps = 1e-6

u = np.zeros((Nt, N))
u[0] = -np.sin(np.pi*x/L)-eps
```

## Left-Hand

```python
A = np.diag(np.ones(N))
A += np.diag(-np.ones(N-1), -1)
A[0,-1] = -1

for i in range(Nt-1):
    u[i+1] = u[i] -  dt/dx*u[i]*(A@u[i])

plt.imshow(u, aspect="auto", cmap="magma")
plt.show()
```

## Right-Hand

```python
A = np.diag(np.ones(N))
A += np.diag(-np.ones(N-1), 1)
A[-1,0] = -1

for i in range(Nt-1):
    u[i+1] = u[i] -  dt/dx*u[i]*(A@u[i])

plt.imshow(u, aspect="auto", cmap="magma")
plt.show()
```

## Central

```python
A = np.diag(np.ones(N-1), 1)
A += np.diag(-np.ones(N-1), -1)
A[-1,0] = 1
A[0,-1] = -1

for i in range(Nt-1):
    u[i+1] = u[i] -  dt/(2*dx)*u[i]*(A@u[i])

plt.imshow(u, aspect="auto", cmap="magma")
plt.show()
```

Left-hand and right-hand are stable much longer than central difference for Burgers Equation. For negative initial conditions the values dont change over all but its unstable for all three cases.
